```python
import pandas as pd
```

```python
df = pd.read_pickle("./data/pre-processed-metrics-dataframe.pkl")
```

```python
df.columns
```

```python
system_properties = [
    "system_type",
    "processor_count", "core_count",
    "network_model",
    "network_bandwidth", "network_latency",
    "min_processor_speed", "max_processor_speed",
    
]

unique_systems_df = df.groupby("system").first()[system_properties]
```

```python
unique_systems_df
```

```python
len(unique_systems_df)
```

```python
gb = unique_systems_df.groupby(["processor_count","system_type","network_model" ])
```

```python
df_systems_summary = gb.agg(lambda x: list(pd.unique(x)))
```

```python
df_systems_summary["# Systems"] = gb.size()
```

```python
df_systems_summary = df_systems_summary.reset_index()
```

```python
df_systems_summary = df_systems_summary[
["# Systems", "processor_count", "system_type", "network_model",
 "core_count",
 "network_bandwidth", "network_latency",
 "min_processor_speed", "max_processor_speed"
]
]
```

```python
df_systems_summary
```

```python
df_systems_summary["system_type"] = (df_systems_summary["system_type"] == "homogeneous")
```

```python
df_systems_summary["system_type"] = df_systems_summary["system_type"].apply(lambda x: "Hom" if x else "Het")
```

```python
# df_systems_summary = df_systems_summary.rename(columns={"system_type":"homogeneous"})
```

```python
def lsort(x):
    x.sort()
    return x

df_systems_summary["core_count"] = df_systems_summary["core_count"].apply(lsort)
```

```python
df_systems_summary.min_processor_speed = df_systems_summary.min_processor_speed.apply(lambda x: x[0])
```

```python
df_systems_summary
```

```python
df_systems_summary.max_processor_speed = df_systems_summary.max_processor_speed.apply(lambda x: x[0])
```

```python
bools = (df_systems_summary.network_model == "ConstantBandwidth")
```

```python
df_systems_summary.loc[bools, "network_model"] = "Const"
df_systems_summary.loc[~bools, "network_model"] = "Shared"
```
