```python
import pandas as pd
```

```python
df = pd.read_pickle("./data/pre-processed-metrics-dataframe.pkl")
```

```python
df.columns
```

```python
dag_properties = [
    "workflow",
    "dag_nodes",
    "dag_edges",
    "dag_density",
    "dag_depth",
    "dag_width",
    "dag_parallelism_degree",
    "CCR",
]

unique_dags_df = df.groupby("dag").first()[dag_properties]
```

```python
gb = unique_dags_df.groupby("workflow")
```

```python
unique_dags_agg_df = gb.agg(lambda x: list(pd.unique(x)))
```

```python
unique_dags_agg_df["Total # of DAGs"] = gb.size()
```

```python
def get_range(x):

    return str(round(min(x), 4)) + " - " + str(round(max(x), 4))
```

```python
unique_dags_agg_df["No. DAG Nodes Range"] = unique_dags_agg_df.dag_nodes.apply(get_range)
```

```python
unique_dags_agg_df["No. DAG Edges Range"] = unique_dags_agg_df.dag_edges.apply(get_range)
```

```python
unique_dags_agg_df["DAG Density Range"] = unique_dags_agg_df.dag_density.apply(get_range)
```

```python
unique_dags_agg_df["DAG Depth Range"] = unique_dags_agg_df.dag_depth.apply(get_range)
```

```python
unique_dags_agg_df["DAG Width Range"] = unique_dags_agg_df.dag_width.apply(get_range)
```

```python
unique_dags_agg_df["DAG Parallelism Range"] = unique_dags_agg_df.dag_parallelism_degree.apply(get_range)
```

```python
unique_dags_agg_df["DAG CCR Range"] = unique_dags_agg_df.CCR.apply(get_range)
```
